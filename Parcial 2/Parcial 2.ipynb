```python
import sys
import numpy as np
import pylab as pl
import random as rnd
```

```python
def main(cantidad):
    archivo = open("texto.txt", "r") #Lectura de archivo de Texto 1
    #archivo = open("ps7_input1.txt", "r") #Lectura de archivo de Texto 2
    #archivo = open("ps7_input2.txt", "r") #Lectura de archivo de Texto 3
    #archivo = open("ps7_input3.txt", "r") #Lectura de archivo de Texto 4
    mat = []
    aux = []
    for l in range(1,cantidad + 1): 
        x = archivo.readline()      #cadena 1
        n = int(archivo.readline()) #tamao cadena 1
        y = archivo.readline()     #cadena 2
        m = int(archivo.readline())  #tamao cadena 2    
        
        print("Cadena 1:",x)
        print("n:",n)
        print(" Cadena 2:",y)
        print("m:",m)
        
        mat = []  
        for i in range(n + 1):
            mat.append([])        
            for j in range(m + 1):
                mat[i].append(0)   
        for i in range(n + 1):
            mat[i][0] = i * 2         
        for j in range(m + 1):
            mat[0][j] = j * 3       
        for i in range(1,n + 1):
            for j in range(1,m + 1):
                if (x[i-1] == y[j - 1]):
                    mat[i][j] = mat[i - 1][j - 1]
                else:
                    mat[i][j] = min(mat[i - 1][j - 1] + 4, 
                                    min(mat[i - 1][j] + 2, 
                                        mat[i][j - 1] + 3))
        aux.append(mat[n][m])    
    return aux
```

```python
n = 1
edi = main(n)
print(edi)
```
